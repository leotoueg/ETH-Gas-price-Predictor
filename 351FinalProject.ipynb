```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

df=pd.read_csv('features.csv')
df.head()
```

Fill any Na value with the mean of that feature

```python

df['addressCount'].fillna((df['addressCount'].mean()), inplace=True)
df['supply'].fillna((df['supply'].mean()), inplace=True)
#df['price'].fillna((df['price'].mean()), inplace=True)
df['blockSize'].fillna((df['blockSize'].mean()), inplace=True)
df['addressCount'].fillna((df['addressCount'].mean()), inplace=True)
df['dailyTransactions'].fillna((df['dailyTransactions'].mean()), inplace=True)
df['NumERC20TokenTransfers'].fillna((df['NumERC20TokenTransfers'].mean()), inplace=True)
df['dailyETHBurnt'].fillna((df['dailyETHBurnt'].mean()), inplace=True)

df.reset_index()
df.head()
```

Presents the mean, standard deviation, and percentiles of each feature. You can compare the standard deviation to the various percentiles to gauge whether there are outliers


Here we will change the price feature from a string to a float

```python
df['price']=df['price'].str.replace(',', '').astype(float)
```

Now we will change the date attribute to a days since conception for use in the model.

```python

df['date'] = pd.to_datetime(df['date'])
df['date_delta'] = (df['date'] - df['date'].min())  / np.timedelta64(1,'D')
df.drop('date',1)
df.describe()
```

Any feature found to show signs of outliers will be trimmed by cutting off values past a certain standard deviation factor.

```python
factor = 2
df['gasPrice']=(df['gasPrice']-df['gasPrice'].mean())/df['gasPrice'].std()
upper_lim = df['gasPrice'].mean () + df['gasPrice'].std () * factor
lower_lim = df['gasPrice'].mean () - df['gasPrice'].std () * factor
clean_data = df[(df['gasPrice'] < upper_lim) & (df['gasPrice'] > lower_lim)]
```

Now that the data has been preprocessed, it is time to construct a model to predict the gas price at any given data. The first step is to split the data into training and validation sets. For the purpose of this project, a 10-90 test-train split will be used.

```python
training_data=clean_data.sample(frac=0.8,random_state=1)
test_data=clean_data.drop(training_data.index)
```

As a start, a simple multiple regression will be performed with the refined data. The results of this regression will help determine whether or not more accurate predictions are possible with more complex models.

```python
from sklearn import linear_model

Y_train=training_data['gasPrice']
X_train=training_data.loc[:,~training_data.columns.isin(['gasPrice','date','networkTransactionFees','ENSRegistration'])]
print(np.any(np.isnan(X_train)))
print(np.all(np.isfinite(X_train)))


regression=linear_model.LinearRegression()
regression.fit(X_train,Y_train)

print(regression.score(X_train,Y_train))
```

A regression score of 0.714 shows promising results for the selected features abilities to predict Ethereum gas fees. We will now use an artificial neural network to try and obtain a better prediciton.

```python
from sklearn.neural_network import MLPRegressor
from sklearn import metrics
plt.style.use('ggplot')
Y_test=test_data['gasPrice']
X_test=test_data.loc[:,~test_data.columns.isin(['gasPrice','date','networkTransactionFees','ENSRegistration'])]

model=MLPRegressor(hidden_layer_sizes=15,max_iter=150)
model.fit(X_train,Y_train)
print(model)
expected_Y=Y_test
predicted_Y = model.predict(X_test)
print(metrics.r2_score(expected_Y, predicted_Y))
print(metrics.mean_squared_error(expected_Y, predicted_Y))
```

The results of the neural network regression turned out to be worse than that of the simple linear regression. Another approach, the Random Forest will be analyzed next

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor()
model.fit(X_train,Y_train)
predicted_Y = model.predict(X_test)
print(metrics.r2_score(expected_Y, predicted_Y))
print(metrics.mean_squared_error(expected_Y,predicted_Y))
print(model.feature_importances_)
```
